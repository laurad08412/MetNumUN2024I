# 3) Example: SymPy’s Symbolic and Numercical Complex Evaluations

Ecuacion compleja y foto de wolfram

a = 2*a1 + I*a2; b = b1^2 + I*b2/3; c = 5*c1 + I*2*c2^2; (a * b) / c


```
from sympy import symbols, I, N, pi, expand

# Definir los símbolos
a1, a2, b1, b2, c1, c2 = symbols("a1 a2 b1 b2 c1 c2", real=True)
a = 2*a1 + I*a2
b = b1**2 + I*(b2/3)
c = 5*c1 + I*2*c2**2

# Calcular la expresión a*b/c y sus expansiones
expression1 = a * b / c
expression2 = expand(a * b / c)
expression3 = expand((a * b) / c)
expression4 = expand(a * (b / c))

# Imprimir resultados
print("a * b / c:", expression1)
print("Expansión de a * b / c:", expression2)
print("Expansión de (a * b) / c:", expression3)
print("Expansión de a * (b / c):", expression4)

# Calcular 1 / (pi + I)
w = 1 / (pi + I)
print('w =', w)

# Convertir w a un número decimal
w = N(w)
print('w =', w)
```
